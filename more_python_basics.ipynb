```
import random
```

```
def factorial(x):
  if (x==0):
    return 1
  return x*factorial(x-1)
```

```
def code_to_debug():
  #import pdb;pdb.set_trace()
  for i in range(10):
    x=random.random()
    factorial(x)
```
